```python
import torch
import numpy as np
from sklearn.datasets import load_digits
from matplotlib import pyplot as plt

digits = load_digits()

plt.figure(figsize=(5,5))
plt.imshow(torch.tensor(digits.data[0]).reshape((8,8)), cmap='binary')

plt.xticks(np.arange(-1/2, 8, 1.))
plt.yticks(np.arange(-1/2, 8, 1.))

plt.grid()
plt.show()
```
