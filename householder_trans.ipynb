```python
import torch
import numpy as np
import matplotlib.pyplot as plt
```

```python
x = torch.tensor([[1],[2]]).to(torch.float32)
```

```python
# = torch.rand((2,1)).to(torch.float32)

#v = torch.tensor([[0],[1]])
v=x
u = v / torch.sqrt(torch.matmul(v.T, v))
```

```python
h = torch.eye(2) - 2 * torch.matmul(v, v.T) / torch.matmul(v.T, v)
```

```python
x_ref = torch.matmul(h, x)
```

```python
plt.arrow(0, 0, x[0,0], x[1,0])
plt.arrow(0, 0, x_ref[0,0], x_ref[1,0])
plt.arrow(0, 0, u[0,0], u[1,0], color="red")
```
