```python
# prompt: 2 dim normal distribution by pytorch distribution library and plot

import torch
import torch.distributions as dist
import matplotlib.pyplot as plt

# define parameters of the distribution
mu = torch.tensor([0., 0.])
sigma = torch.tensor([[1., 0.], [0., 1.]])

# create a multivariate normal distribution
mvn = dist.MultivariateNormal(mu, sigma)

# sample from the distribution
x = mvn.sample((1000,))

# plot the samples
plt.scatter(x[:, 0], x[:, 1])
plt.show()
```

```python
# prompt: how to make contour plot in 2 dim normal distribution

import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import multivariate_normal

# Define the parameters of the distribution
mu = np.array([0, 0])
sigma = np.array([[1, 0], [0, 1]])

# Create a grid of points to evaluate the distribution at
x = np.linspace(-3, 3, 100)
y = np.linspace(-3, 3, 100)
X, Y = np.meshgrid(x, y)

# Evaluate the distribution at each point
Z = multivariate_normal.pdf(np.array([X, Y]).T, mean=mu, cov=sigma)

plt.figure(figsize=(4, 4))
# Plot the contour plot
plt.contourf(X, Y, Z)
plt.xlabel('x')
plt.ylabel('y')
plt.show()
```
